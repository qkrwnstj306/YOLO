```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from torch.utils.data import Dataset, DataLoader

import matplotlib.pyplot as plt
from numpy import asarray
```

```python
torch.manual_seed(123)
```

```python
transforms_data = transforms.Compose([
    transforms.ToTensor(), 
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(180, fill=(30,150,45), expand=True),
    transforms.Resize((224, 224)),
])
```

```python
class CustomDataset(Dataset):
    def __init__(self, x_data, transforms_data):
        self.x_data = [asarray(x_data)] # data가 하나라 감싸줌
        self.transforms = transforms_data
    def __len__(self):
        return len(self.x_data)

    def __getitem__(self, index):
        self.x_data[index] = self.transforms(self.x_data[index])
        return self.x_data[index]
```

```python
import PIL

img = PIL.Image.open('./cat0.jpg')
print(img.size)
print(transforms.ToTensor()(img).size())
print(asarray(img).shape)
```

```python
dataset = CustomDataset(img, transforms_data)
print(dataset.__len__())
dataloader = DataLoader(dataset, batch_size = 2, shuffle = True)

for idx, image in enumerate(dataloader):
    print(image.shape)
    
    image = image.squeeze(0).permute(1,2,0)
    
    plt.imshow(image)
plt.show()
```
